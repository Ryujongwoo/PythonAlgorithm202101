```python
# 노래 제목 크롤링
url = 'https://www.genie.co.kr/chart/top200?ditc=D&ymd=20210119&hh=21&rtm=Y&pg='
for i in range(1, 2):
    targetSite = '{}{}'.format(url, i)
    # print(targerSite)
    h = {'User-agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64; Trident/7.0; rv:11.0) like Gecko'}
    request = requests.get(targetSite, headers = h)
    html = request.text
    soup = BeautifulSoup(html)
    # print(soup)
    
    titles = soup.findAll('a', {'class': 'title ellipsis'})
    # print(titles)
    # print(len(titles))
    
    for title in titles:
        # print(title.text.strip())
        # print(title.text.split('\n')[8].strip())
        print(title.text.strip().split('\n')[0])
```

```python
# 아티스트 크롤링
url = 'https://www.genie.co.kr/chart/top200?ditc=D&ymd=20210119&hh=21&rtm=Y&pg='
for i in range(1, 2):
    targetSite = '{}{}'.format(url, i)
    # print(targerSite)
    h = {'User-agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64; Trident/7.0; rv:11.0) like Gecko'}
    request = requests.get(targetSite, headers = h)
    html = request.text
    soup = BeautifulSoup(html)
    # print(soup)
    
    artists = soup.findAll('a', {'class': 'artist ellipsis'})
    # print(artists)
    # print(len(artists))
    
    for artist in artists[5:]:
        # print(artist.text.strip())
        # print(artist.text.split('\n')[0])
        print(artist.text.strip().split('\n')[0])
```

```python
# 크롤링한 내용을 리스트에 저장한다.
genie = []
for i in range(1, 5):
    targetSite = '{}{}'.format(url, i)
    h = {'User-agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64; Trident/7.0; rv:11.0) like Gecko'}
    request = requests.get(targetSite, headers = h)
    time.sleep(1)
    html = request.text
    soup = BeautifulSoup(html)
    
    titles = soup.findAll('a', {'class': 'title ellipsis'})
    artists = soup.findAll('a', {'class': 'artist ellipsis'})
    
    for i in range(50):
        genie.append(
            [titles[i].text.strip().split('\n')[0], artists[i + 5].text.strip().split('\n')[0]]
        )
# print(len(genie))
# print(genie)
```

```python
print(dt.now())
for i in range(200):
    print('{0:3d}위 {1} - {2}'.format(i + 1, genie[i][1], genie[i][0]))
```
