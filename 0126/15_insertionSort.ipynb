```python
# 이미 정렬된 파일에 새로운 데이터를 위치에 맞게 삽입한다.
for i in range(1, 5):
    for j in range(i - 1, -1, -1):
        print('i = {}, j = {}'.format(i, j))
```

```python
sortData = [8, 3, 4, 9, 1]
for i in range(1, len(sortData)): # 삽입할 새 데이터의 위치
    # 이미 정렬된 파일에 새 데이터가 삽입될 위치를 찾기 위해서 새 데이터와 인덱스를 각각 key, index 변수에 저장한다.
    key = sortData[i]
    index = i
    # 새 데이터가 삽입될 위치를 찾는다.
    for j in range(i - 1, -1, -1):
        if sortData[j] > key:
            # 정렬된 파일의 데이터가 크기 때문에 정될된 파일의 데이터를 인덱스가 1큰 위치로 이동시키과 삽입할 데이터의
            # 인덱스를 1감소 시킨다.
            sortData[j + 1] = sortData[j]
            index -= 1
        else:
            # 삽입될 데이터가 이미 정렬된 파일의 데이터보다 크거나 같다는 의미이므로 이미 정렬된 데이터 파일의 앞 부분
            # 데이터와 더 이상 비교할 필요가 없으므로 반복할 필요가 없어진다. => 반복을 탈출시킨다.
            break
        # ===== if 끝
    # ===== for j 끝 => 회전 끝
    # 새 데이터를 정렬된 파일의 위치에 맞게 삽입한다.
    sortData[index] = key
    print('{}회전 결과 : {}'.format(i, sortData))
# ===== for i 끝
# print('정렬 결과 : {}'.format(sortData))
```
