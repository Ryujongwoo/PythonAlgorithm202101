```python
sortData = [8, 3, 4, 9, 1]
for i in range(1, len(sortData)): #    
    #              key, index  .
    key = sortData[i]
    index = i
    #     .
    for j in range(i - 1, -1, -1):
        if sortData[j] > key:
            #          1    
            #  1 .
            sortData[j + 1] = sortData[j]
            index -= 1
        else:
            #               
            #         . =>  .
            break
        # ===== if 
    # ===== for j  =>  
    #       .
    sortData[index] = key
    print('{}  : {}'.format(i, sortData))
# ===== for i 
# print('  : {}'.format(sortData))
```
